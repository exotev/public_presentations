```python
import geocoder
import geopy
from geopy.geocoders import Nominatim
g = geocoder.ip('me')
print("My Location is:")
print(g.latlng)
```

```python
geolocator = Nominatim(user_agent="location_app")
```

```python
user_location = geolocator.geocode('Mauritania, Nouakchott')
```

```python
user_location.raw
```

```python
villes = "Nouakchott Nouadhibou Rosso Adel Bagrou Boghé Kiffa Zouérat édi Boû Gâdoûm Boutilimit Atâr Bareine Ghabou Hamoud Mâl N'Beike Gouraye Timbédra Magta Lahjar Guérou Soudoud Sélibaby Voum Legleite Legceïba Boû Steïlé Sangrave Tintane Néma Aleg Tidjikja Walata Ayoûn el-Atroûs Bababé Aqjawajat Chagar".split(" ")
```

```python
gps_data = {}
for v in villes:
    loc = geolocator.geocode("Mauritania, " + v)
    gps_data[v] = {'ville' : v}
    if(loc is not None):
        gps_data[v]['latitude'] = loc.point.latitude
        gps_data[v]['longitude'] = loc.point.longitude
        gps_data[v]['altitude'] = loc.point.altitude
    print(gps_data[v])
```

```python
import pandas as pd
gps_df = pd.DataFrame(gps_data)
```
